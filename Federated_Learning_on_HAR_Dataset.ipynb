```python
# here we are labelling the categorical valuee with numerical values
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.fit_transform(y_test)
```

```python
import numpy as np
import tensorflow as tf
import keras
from keras import layers

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Define the client class
class Client:
    def __init__(self, client_id, data):
        self.client_id = client_id
        self.data = data
        self.model = self.initialize_model()

    def initialize_model(self): # 2 layer
        model = keras.Sequential([
            layers.Dense(64, activation='relu'),
            layers.Dense(6, activation='softmax')
        ])
        model.compile(optimizer='adam',
                      loss='sparse_categorical_crossentropy',
                      metrics=['accuracy'])
        return model

    def train(self, num_epochs=5, batch_size=32):# train the model on the clients data
        x_train, y_train = self.data
        self.model.fit(x_train, y_train, epochs=num_epochs, batch_size=batch_size)

    def get_model_weights(self): # give the current model weight
        return self.model.get_weights()

    def set_model_weights(self, weights):
        self.model.set_weights(weights)

#Fed learning setup
num_clients = 10
num_epochs = 5
batch_size = 32

# Split the data among clients
client_data_size = len(x_train) // num_clients # data size for each client
# Slicing of data for each ckient from the data set
'''
the first client will get the first 'client_data_size' samples from x_train and y_train,
the second client will get the next 'client_data_size' samples, and so on.
'''
client_data = [(x_train[i*client_data_size:(i+1)*client_data_size],
                y_train[i*client_data_size:(i+1)*client_data_size])
               for i in range(num_clients)]

# Create clients
clients = [Client(client_id, data) for client_id, data in enumerate(client_data)]


def average_weights(weights_list):
    # Assume weights_list is a list of lists of model weights
    # Calculate element-wise average of all weights
    aggregated_weights = [np.mean(w, axis=0) for w in zip(*weights_list)]
    return aggregated_weights

# Federated learning loop
for epoch in range(num_epochs):
    print(f"Epoch {epoch+1}/{num_epochs}")
    all_model_weights = []  # List to store model weights from all clients
    for client in clients:
        client.train(num_epochs=1, batch_size=batch_size)
        model_weights = client.get_model_weights()
        all_model_weights.append(model_weights)

    # Aggregate model weights (e.g., by averaging)
    aggregated_weights = average_weights(all_model_weights)

    # Distribute aggregated weights back to the clients
    for client in clients:
        client.set_model_weights(aggregated_weights)
'''



1. Client Class: This class represents a client in the federated learning setup. Each client has its own data and model. The class has several methods:
    - `__init__`: Initializes the client with a unique ID, data, and a model.
    - `initialize_model`: Defines and compiles a simple Keras model with two layers.
    - `train`: Trains the model on the client's data.
    - `get_model_weights`: Returns the current weights of the model.
    - `set_model_weights`: Sets the model's weights.

2. Federated Learning Setup: The code then sets up the federated learning environment. It first defines the number of clients, epochs,
and batch size. It then splits the training data among the clients and initializes them.

3. average_weights Function: This function takes a list of model weights (each item in the list is a set of weights from one client)
 and returns the average of these weights. This is used to aggregate the model updates from all clients.

4. Federated Learning Loop: This is the main loop where federated learning happens. For each epoch, it does the following:
    - Each client trains its model for one epoch and then the model weights are collected.
    - The weights from all clients are averaged using the `average_weights` function. This is the aggregated update that is sent back to the server.
    - The aggregated weights are then distributed back to all the clients, where they are used to update the clients' models.

```

```python
# Global Model
'''
This is a common practice in federated learning where after the learning process,
a global model is created using the aggregated weights. This model can then be used for further tasks like prediction, evaluation,
'''
def create_aggregated_model(aggregated_weights, input_shape):
    model = keras.Sequential([
        layers.Input(shape=input_shape),
        layers.Dense(64, activation='relu'),
        layers.Dense(6, activation='softmax')
    ])
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    model.set_weights(aggregated_weights)
    return model

# Create the aggregated model using the aggregated weights
input_shape = x_train.shape[1:]
aggregated_model = create_aggregated_model(aggregated_weights, input_shape)
```

```python
# Evaluate the Global model on the test set
test_loss, test_accuracy = aggregated_model.evaluate(x_test, y_test)

print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_accuracy}")
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Define activity labels
activity_labels = ['STANDING', 'SITTING', 'LAYING', 'WALKING', 'WALKING_DOWNSTAIRS', 'WALKING_UPSTAIRS']

# Get predictions for the test data using the aggregated model
y_pred = np.argmax(aggregated_model.predict(x_test), axis=1)

# Calculate the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)


# Plot the confusion matrix
plt.figure(figsize=(10, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.xticks(ticks=np.arange(6) + 0.5, labels=activity_labels, rotation=30)
plt.yticks(ticks=np.arange(6) + 0.5, labels=activity_labels, rotation=30)
plt.show()
```
